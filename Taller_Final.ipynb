

#  PBM vs EDAD

Variables necesarias:

EDAD: edad de la persona

PBM: Puntaje básico de matrícula

```python
# Mostramos un vistazo rápido para verificar que las columnas existen
print(data[['EDAD', 'PBM']].head())
```

```python
# Seleccionamos solo las columnas necesarias
data_EDAD_PBM = data[['EDAD', 'PBM']].copy() # Changed 'df' to 'data'

# Eliminamos filas con valores faltantes (NaN)
data_EDAD_PBM.dropna(inplace=True)

# Filtramos edades realistas (evitar outliers como <16 o >60)
data_EDAD_PBM = data_EDAD_PBM[(data_EDAD_PBM['EDAD'] >= 16) & (data_EDAD_PBM['EDAD'] <= 60)] # Changed 'edad' to 'EDAD'

# También limitamos PBM a un rango lógico (ejemplo: 0 a 1000)
data_EDAD_PBM = data_EDAD_PBM[(data_EDAD_PBM['PBM'] > 0) & (data_EDAD_PBM['PBM'] < 1000)] # Changed 'pbm' to 'PBM'
```

```python
#Visualización 1: Gráfico de dispersión

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_EDAD_PBM, x='EDAD', y='PBM', alpha=0.3, color='blue')
plt.title("Relación entre Edad y PBM", fontsize=14)
plt.xlabel("EDAD")
plt.ylabel("PBM")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Visualización 2: Boxplot por rangos de edad
# Creamos una nueva columna con rangos de edad

data_EDAD_PBM['rango_edad'] = pd.cut(data_EDAD_PBM['EDAD'], bins=[15, 20, 25, 30, 35, 40, 50, 60],
                                   labels=["16-20", "21-25", "26-30", "31-35", "36-40", "41-50", "51-60"])

plt.figure(figsize=(10, 6))
sns.boxplot(data=data_EDAD_PBM, x='rango_edad', y='PBM', palette='viridis') # Changed 'pbm' to 'PBM'
plt.title("Distribución del PBM por Rango de Edad", fontsize=14)
plt.xlabel("Rango de Edad")
plt.ylabel("PBM")
plt.tight_layout()
plt.show()
```

# PBM por procedencia geográfica
Variables que usaremos:

PBM: Puntaje básico de matrícula

DEP_PROC: Departamento de procedencia

CIU_PROC: Ciudad de procedencia

LAT_CIU_PROC, LON_CIU_PROC: Coordenadas geográficas del municipio de procedencia

```python
# Para mapas
import geopandas as gpd
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
# Nos quedamos con las columnas necesarias
data_geo = data[['PBM', 'DEP_PROC', 'CIU_PROC', 'LAT_CIU_PROC', 'LON_CIU_PROC']].copy()

# Limpiamos datos
data_geo.dropna(inplace=True)
data_geo = data_geo[(data_geo['PBM'] > 0) & (data_geo['PBM'] < 1000)]
```

```python

#PBM promedio por Departamento (Top 15)

# Agrupar por departamento y calcular promedio de PBM
pbm_por_departamento = data_geo.groupby('DEP_PROC')['PBM'].mean().sort_values(ascending=False).head(15)

# Gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x=pbm_por_departamento.values, y=pbm_por_departamento.index, palette='rocket') # Corrected the variable name
plt.title("Top 15 Departamentos con Mayor PBM Promedio", fontsize=14)
plt.xlabel("PBM Promedio")
plt.ylabel("Departamento de Procedencia")
plt.tight_layout()
plt.show()
```

```python
# Agrupar por ciudad y coordenadas para obtener el PBM promedio por ciudad
pbm_por_ciudad = data_geo.groupby(['CIU_PROC', 'LAT_CIU_PROC', 'LON_CIU_PROC'])['PBM'].mean().reset_index() # Changed df_geo to data_geo and corrected column names

# Mapa con puntos
plt.figure(figsize=(10, 8))
sc = plt.scatter(
    x=pbm_por_ciudad['LON_CIU_PROC'],  # Changed lon_ciu_proc to LON_CIU_PROC
    y=pbm_por_ciudad['LAT_CIU_PROC'],  # Changed lat_ciu_proc to LAT_CIU_PROC
    c=pbm_por_ciudad['PBM'],           # Changed pbm to PBM
    cmap='viridis',
    s=50,
    alpha=0.7
)

plt.colorbar(sc, label='PBM Promedio')
plt.title('Mapa de Calor del PBM por Ciudad de Procedencia')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Promedio de PBM por año y semestre
data['PERIODO'] = data['YEAR'].astype(str) + "-" + data['SEMESTRE'].astype(str)

pbm_tiempo = data.groupby('PERIODO')['PBM'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=pbm_tiempo, x='PERIODO', y='PBM', marker='o', color='teal')
plt.xticks(rotation=45)
plt.title("Evolución del PBM promedio por periodo (año-semestre)")
plt.xlabel("Periodo")
plt.ylabel("PBM Promedio")
plt.tight_layout()
plt.show()
```

```python
#Evolución de proporciones por sexo en el tiempo
sexo_tiempo = data.groupby(['PERIODO', 'SEXO']).size().reset_index(name='conteo')
total_periodo = sexo_tiempo.groupby('PERIODO')['conteo'].transform('sum')
sexo_tiempo['PROPORCION'] = sexo_tiempo['conteo'] / total_periodo

plt.figure(figsize=(12, 6))
sns.lineplot(data=sexo_tiempo, x='PERIODO', y='PROPORCION', hue='SEXO', marker='o')
plt.title("Proporción de estudiantes por sexo a lo largo del tiempo")
plt.xlabel("Periodo")
plt.ylabel("Proporción")
plt.xticks(rotation=45)
plt.legend(title="Sexo")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd # Make sure pandas is imported

#Nº de Outliers usando rango IQR
outliers={}

# Get numeric columns before the loop
num_cols = data.select_dtypes(include='number').columns

for col in num_cols:
  Q1 = np.percentile(data[col], 25)
  Q3 = np.percentile(data[col], 75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers[col]= (data[col] > upper_bound).sum() + (data[col]< lower_bound).sum()

print(outliers)
```

```python
#Nº de Outliers usando 2 std
outliers={}
for col in num_cols:
  mean=data[col].mean()
  std=data[col].std()

  outliers[col] = (data[col] > (mean + 2 *std)).sum() + (data[col] < (mean - 2 *std)).sum()

print(outliers)
```

```python

#Nº de Outliers usando 2 std
data = pd.DataFrame(data)

# Identificar columnas numéricas
num_cols = data.select_dtypes(include=['float64', 'int64']).columns

# Inicializar diccionario para almacenar outliers y desviaciones estándar
outliers = {}
std_devs = {}

for col in num_cols:
    # Calcular media y desviación estándar para la columna actual
    mean = data[col].mean()
    std = data[col].std()

    # Calcular límites superior e inferior para la detección de outliers
    upper_bound = mean + 2 * std
    lower_bound = mean - 2 * std

    # Identificar y contar outliers para la columna actual
    num_outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)].shape[0]

    # Almacenar el número de outliers y la desviación estándar en los diccionarios
    outliers[col] = num_outliers
    std_devs[col] = std

# Mostrar el conteo de outliers
print(outliers)

# Crear un DataFrame para la visualización
outlier_data = pd.DataFrame({
    'Outliers': outliers,
    'Standard Deviation': std_devs
})

# Graficar
fig, ax1 = plt.subplots(figsize=(12, 6))
outlier_data.plot(kind='bar', ax=ax1, color='lightblue', alpha=0.7)  # Plot on the primary axis
ax1.set_title('Número de Outliers y Desviación Estándar por Columna')
ax1.set_xlabel('Columnas')
ax1.set_ylabel('Número de Outliers')
ax1.tick_params(axis='x', rotation=45) # Rotate x-axis labels for better readability
ax1.grid()

# Create a secondary y-axis
ax2 = ax1.twinx()
ax2.set_ylabel('Desviación Estándar')

plt.show()
```
